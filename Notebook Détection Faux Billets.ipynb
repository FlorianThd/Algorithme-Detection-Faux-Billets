# <font color="#8333ff">R√©gression lin√©aire multiple</font>

## Cr√©ation du mod√®le

Afin de combler nos donn√©es manquantes, nous allons tenter d'estimer les donn√©es manquantes 'margin_low' **√† partir des donn√©es des 5 autres variables quantitatives**.</br> <br> On appellera :<br> <br> --> **x_var** le DataFrame comprenant les 5 variables quantitatives que l'on conna√Æt, <br> et <br> --> **y_var** la variable "margin_low" pour laquelle on cherche √† recr√©er les donn√©es manquantes.

```python
# nos variables quantitatives que l'on conna√Æt
var_connues = ["diagonal","height_left","height_right","margin_up","length"]

# nos deux dataframes **variables connues** / **variable √† pr√©dire**
x_var = bil_nonan[var_connues]
y_var = bil_nonan["margin_low"]

# on cr√©e une variable "lr" √† partir d'un mod√®le lin√©aire "LinearRegression"
lr = LinearRegression()

# on entra√Æne notre variable lr avec les donn√©es x_var, y_var
lr.fit(x_var,y_var)

# sur la base de cet entra√Ænement on g√©n√©re une pr√©diction y_predict 
y_predict = lr.predict(x_var)
```

## Le R¬≤ ou coefficient de d√©termination

Notre mod√®le de pr√©diction est naturellement imparfait.

Le coefficient de d√©termination, appel√© R¬≤, repr√©sente la part de la variation de la variable Y (dans notre cas, margin_low) expliqu√©e par notre mod√®le.

Il est le rapport entre la SCE
(Somme des Carr√©s Expliqu√©e => traduit la part de la variation expliqu√©e par le mod√®le.)
et la SCT (Somme des Carr√©s Totale=> traduit la variation totale de Y).

Plus le R¬≤ est √©lev√©, plus le mod√®le explique une large part de la variation de Y.


Voyons maintenant si notre mod√®le r√©alise des pr√©dictions fiables, gr√¢ce au co√©fficient de d√©termination :

```python
lr.score(x_var, y_var)
```

**Le R¬≤ est malheureusement tr√®s bas. Le mod√®le explique une faible part de la variation de notre varibale d√©pendante.<br> Nous allons tenter de trouver les variables qui contribuent le moins et le plus √† la pr√©diction de notre variable d√©pendante.**

```python
var_qti = bil_nonan.columns.drop('is_genuine').tolist() # cr√©ation de la liste de toutes les variables quantitatives
display(var_qti)
display(var_connues) # en comparaison, var_connues donne les variables quanti hormis 'margin_low'
```

## Test de la significativit√© de chaque variable

```python
reg_multi = log_reg = smf.ols('margin_low~diagonal+height_left+height_right+margin_up+length', data=bil_nonan).fit()
print(reg_multi.summary())

# Tous les param√®tres sont significatifs.( P>|t| quasi √©gal √† 0)
```

## Test de la normalit√© des r√©sidus

Notre mod√®le √©tant imparfait,
essayons de repr√©senter la "part d'erreur" de cette pr√©diction,<br>
c'est √† dire notre pr√©diction moins les donn√©es r√©elles, observ√©es.

```python
display(y_var.values) # nos donn√©es r√©elles
display(np.round(y_predict,2)) # nos donn√©es pr√©dites
residus = y_var - y_predict # calculs de la diff√©rence "donn√©es r√©elles" - "donn√©es pr√©dites"
display(residus.values)
plt.hist(residus)
plt.show()
```

```python
shapiro(reg_multi.resid)
```

Ici, l'hypoth√®se de normalit√© est remise en cause (p-value =~ 0 < 0.05).<br> Cela √©tant nous avons un √©chantillon qui a une grande quantit√© d'individus (tr√®s sup√©rieur √† 30), et on observe que graphiquement, m√™me si cela n'est pas parfait, on obtient une distribution qui se rapproche d'une distribution normale.

## Test de l'homo-sc√©dasticit√©

On peut √©galement tester l‚Äôhomosc√©dasticit√© (c'est-√†-dire la constance de la variance) des r√©sidus (des erreurs) :

(√âtymologie de ¬´ sc√©dastique ¬ª : du grec ancien œÉŒ∫ŒµŒ¥Œ±œÉœÑŒπŒ∫œåœÇ, skedastikos (¬´ dispersif, dissipatif ¬ª)

Hypoth√®se ùêª0 : il y a homosc√©dasticit√©<br>
Hypoth√®se ùêªA : il y a h√©t√©rosc√©dasticit√©

```python
_, pval, __, f_pval = smd.het_breuschpagan(reg_multi.resid, reg_multi.model.exog)
print('p value test Breusch Pagan:', pval)
```

La p-valeur ici est quasiment √©gale √† 0, donc tr√®s inf√©rieure √† 5% : on doit rejetter l'hypoth√®se ùêª0 selon laquelle les variances sont constantes (l'hypoth√®se d‚Äôhomosc√©dasticit√©).
Potentiellement, notre mod√®le sera peu fiable.

## Variance Inflation Factor (Multicolin√©arit√©)

VIF = l'√©quivalent de 1 / (1-R¬≤)
Plus notre score VIF est √©lev√© pour une variable donn√©e, plus cette variable est corr√©l√©e √† d'autre. Elle sera donc redondante, et peut interf√©rer avec notre mod√®le.

```python
features = bil.columns.drop(["is_genuine",'margin_low'])
data = reg_multi.model.exog
df = pd.DataFrame(data)

vif = pd.DataFrame()
vif['var'] = features
vif['VIF'] = [variance_inflation_factor(data, i) for i in range(1,data.shape[1])]
display(vif)
```

Les scores VIF obtenus sont tr√®s inf√©rieurs √† 10, on peut donc d√©duire qu'ici il n'y a pas de probl√®me majeur de colin√©arit√©.

# <font color="#8333ff">K-Means</font>
ou algorithme de classification non-supervis√©e

```python
# cr√©ation d'un Dataframe sans l'information vrai/faux billet
bil_km = bil2.drop("is_genuine",axis=1)
bil_km.head()
```

## D√©termination du nombre de clusters

```python
kmeans_kwargs = {"init":"random","n_init":10,"max_iter":300,"random_state":42} # cr√©ation des arguments de la fonction

# somme des erreurs au carr√© en fonction du nombre de clusters
sse = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(bil_km)
    sse.append(kmeans.inertia_)
    
sse # visualisation des r√©sultats
```

```python
plt.style.use("seaborn")
plt.plot(range(1, 10), sse)

plt.title("Graphique du coude")

plt.xlabel("Nombre de clusters")
plt.xticks(range(1,10))

plt.ylabel("SSE (Somme des carr√©es des erreurs)")
plt.yticks(sse)

plt.show()
```

```python
# visuellement on pourrait h√©siter entre 2 ou 3 clusters
# la m√©thode KneeLocator nous indique que le nombre optimal de cluster est de 2 :

kl = KneeLocator(range(1, 10), sse, curve="convex", direction="decreasing")
kl.elbow
```

## Cr√©ation du mod√®le

```python
# cr√©ation d'une variable kmeans2 pour 2 clusters √† partir de notre mod√®le KMeans
kmeans2 = KMeans(init='random', n_clusters=2, n_init=10, max_iter =300,random_state=42)

# on entra√Æne notre variable kmeans2 avec les donn√©es bil_km_cr
kmeans2.fit(bil_km)
```

```python
# on r√©cup√®re les num√©ros des labels dans notre variable cl_kmeans
cl_kmeans = kmeans2.labels_

# on cr√©e un dataframe bil_km, rattachant chaque individu de bil2 √† son num√©ro de cluster
bil_clusters = bil2.assign(cl_kmeans=kmeans2.labels_)

# ce qui donne
display(bil_clusters)
```

```python
Y_PRED = ~kmeans2.predict(bil_km).astype(bool)
```

```python
display(Y_PRED)
```

```python
Y_REAL = bil2["is_genuine"]
```

```python
cm_km = confusion_matrix(Y_REAL, Y_PRED)
confusion_matrix_kmeans_display = ConfusionMatrixDisplay(confusion_matrix = cm_km, display_labels = [False, True])
confusion_matrix_kmeans_display.plot()
plt.show()
```

## K-Means

```python
# matrice de confusion

print("Voici notre matrice de confusion pour la m√©thode K-Means : \n\n"+str(cm_km))

fp_km = cm_km[0,1]
fn_km = cm_km[1,0]

total_km = cm_km[0,0] + cm_km[0,1] + cm_km[1,0] + cm_km[1,1]
precision_kmeans = (100 - round((((fp_km+fn_km)/total_km)*100),2))
# m√©thode score
print("\nLa pr√©cision du mod√®le est de : "+str(precision_kmeans)+" %")
print("\nSur un √©chantillon de "+str(total_km)+" occurences :")
print("\nIl y a eu "+str(fp_km)+" faux positif(s)")
print("\nIl y a eu "+str(fn_km)+" faux n√©gatif(s)")
```
